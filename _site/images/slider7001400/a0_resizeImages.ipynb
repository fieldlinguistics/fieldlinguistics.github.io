```python
# reduce sizes of images
import os
from PIL import Image
path = '/Users/mbw/Dropbox (BIDMC Dropbox Team)/0_Work/Webpage/bdsp-core.github.io/images/slider7001400/'
dir_list = os.listdir(path)
print(dir_list)
```

```python
for x in dir_list:
    print('********')
    try: 
        print(x)

        im = Image.open(x)

        # The file format of the source file.
        print(im.format) # Output: JPEG

        # The pixel format used by the image. Typical values are "1", "L", "RGB", or "CMYK."
        print(im.mode) # Output: RGB

        # Image size, in pixels. The size is given as a 2-tuple (width, height).
        print(im.size) # Output: (1920, 1280)

        # Colour palette table, if any.
        print(im.palette) # Output: None
   
        mywidth = 1000
        wpercent = (mywidth/float(im.size[0]))
        hsize = int((float(im.size[1])*float(wpercent)))
        im2 = im.resize((mywidth,hsize), Image.ANTIALIAS)
        imName = x + '_resized.jpg'
        print(imName)
        im2.save(imName)
#         im2.show()

        
    except:
        pass
```
